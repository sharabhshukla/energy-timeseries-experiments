```python
import numpy as np
import pandas as pd
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
from copy import deepcopy
```

```python
raw_data = pd.read_csv('../data/energydata_complete_transformed.csv')
```

```python
#Add datetime attributes
raw_data['hour'] = raw_data['date'].dt.hour
raw_data['dayofweek'] = raw_data['date'].dt.dayofweek
raw_data['month'] = raw_data['date'].dt.month
```

```python
raw_data
```

```python
train, test = raw_data[0:-1000], raw_data[-1000:]
```

```python
X_train = train.drop(columns=['Appliances'], axis=1)
Y_train = train['Appliances']

X_test = test.drop(columns=['Appliances'], axis=1)
Y_test = test['Appliances']
```

```python
lgb_model = LGBMRegressor(n_estimators = 500)
```

```python
lgb_model.fit(X_train, Y_train)
```

```python
pred = lgb_model.predict(X_test)
```

```python
import plotly.express as px

result_df = pd.DataFrame({'pred': pred, 'test': Y_test})

px.line(result_df, y=['pred', 'test'])
```
